```python
import urllib
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://en.wikipedia.org/wiki/List_of_largest_companies_in_Africa_by_revenue"
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
soup = BeautifulSoup(page.text, 'html')
print(soup)
```

```python
soup.find_all('table')[1]
```

```python
table = soup.find('table', class_='wikitable sortable')
world_table_titles = table.find_all('th')
```

```python
world_table_titles
```

```python
world_table_titles=[title.text.strip() for title in world_table_titles]
print(world_table_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns=world_table_titles)
df
```

```python
column_data = table.findAll('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    
    length = len(df)
    df.loc[length] = individual_row_data
```
